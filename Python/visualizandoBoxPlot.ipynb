```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('trees.csv')
dados.head()
```

```python
plt.boxplot(dados.Volume, vert = False, showfliers= False, notch = True, patch_artist=True)
plt.title = 'Árvores'
plt.xlabel = 'Dados'
```

```python
## dados por linhas
plt.boxplot(dados)
plt.title = 'Árvore'
plt.xlabel = 'Dados'
```

```python
## gerando tres boxplot cada um mostrando informações diferentes
plt.boxplot(dados.Height, vert=False)
plt.boxplot(dados.Volume, vert=False)
plt.boxplot(dados.Girth, vert=False)
plt.title = 'Árvores'
plt.xlabel = 'Dados'
```
