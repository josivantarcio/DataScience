```python
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

```python
## criacao de uma variavel com dados de uma distribuição normal com a função rvs (1000 elementos)
dados = norm.rvs(size = 1000)
dados
```

```python
## histograma
plt.hist(dados,bins=20)
plt.title('dados')
```

```python
## gerando de gráfico para verificar se o grafico é normal
fig, ax  = plt.subplots()
stats.probplot(dados, fit=True, plot=ax)
plt.show()
```

```python
## Teste de Shapiro
### segundo argumento é o valor de p, não há como rejeitar a hipotese de nula
### acima de 0.05
stats.shapiro(dados)
```

```python
# Dados não normais
```

```python
dados2 = skewnorm.rvs(4, size=1000)
```

```python
## histograma
plt.hist(dados2, bins=20)
plt.title('dados2')
```

```python
## gerando de gráfico para verificar se o grafico é normal
fig, ax  = plt.subplots()
stats.probplot(dados2, fit=True, plot=ax)
plt.show()
```

```python
## Teste de Shapiro
### segundo argumento é o valor de p, não há como rejeitar a hipotese de nula
### acima de 0.05
stats.shapiro(dados2)
```
