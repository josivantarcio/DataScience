```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base = pd.read_csv('trees.csv', sep=',')
base.head()
```

```python
## histograma com 10 divisoes (bins) e somente para a primeira atributo da base de dados
plt.hist(base.iloc[:,1], bins=10)
```

```python
## histograma com a linha de densidade de frequencia, com 6 divisões(bins)
## kde = linha de densidade
sns.histplot(base.iloc[:,1], kde=False, bins = 6, color='gray').set(title='Árvore')
```

```python
## desidade, metodo kdeplot para densidade
sns.kdeplot(base.iloc[:,1]).set(title='Árvore')
```

```python
## histograma e densidade
sns.histplot(base.iloc[:,1], kde=True, bins = 6)
```
