```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## carregando a base de dados
dados = pd.read_csv('trees.csv', sep=',')
dados.head()
```

```python
## grafico de Dispersao - considerando o volume e a dispersão - Compara duas variaveis
plt.scatter(x = dados.Girth, y = dados.Volume, color = 'blue', marker='*', facecolor = 'none')
plt.title = 'Árvores'
plt.xlabel = 'Circunferencia'
plt.ylabel = 'Volume'
```

```python
## grafico de linha considerando o volume e o atributo 'girth' - sem maker
plt.plot(dados.Girth, dados.Volume)
plt.title = 'Árvores'
plt.xlabel = 'circunferencia'
plt.ylabel = 'volume'
```

```python
## gráfico de dispersão com 'afastamento' de dados (jitter)
## fit_reg - linha de tendencia
sns.regplot(x=dados.Girth, y=dados.Volume, data = dados, x_jitter= 0.3, fit_reg=False)
```
